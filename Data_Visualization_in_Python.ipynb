```python
# import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = {
    "Year": [2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024],
    "Sales": [220, 250, 270, 300, 310, 280, 350, 370, 400, 420],
    "Profit": [60, 70, 80, 90, 100, 95, 110, 120, 130, 140],
    "Category": ["A", "B", "A", "C", "B", "A", "C", "B", "C", "A"],
    "Growth": [0.10, 0.12, 0.15, 0.08, 0.20, 0.18, 0.14, 0.22, 0.19, 0.16]
}

df = pd.DataFrame(data)
print(df)
```

```python
# Matplotlib visualization
```

**LINE PLOT**

```python
plt.plot(df['Year'], df['Sales'], marker="o", color="blue")
plt.title("Sales Over Years (Matplotlib)")
plt.xlabel('Year')
plt.ylabel('Sales')
plt.show()
```

BAR CHART

```python
# Bar chart(Profit by year)
plt.bar(df["Year"], df["Profit"], color="green")
plt.title("Profit by Year (Matplotlib)")
plt.xlabel("Year")
plt.ylabel("Profit")
plt.show()
```

SCATTER PLOT

```python
# scatter plot (sales vs profit)
plt.scatter(df["Sales"], df["Profit"], color="red")
plt.title("Sales vs Profit (Matplotlib)")
plt.xlabel("Sales")
plt.ylabel("Profit")
plt.show()
```

PIE CHART

```python
# PIE CHART  (category Distrubution)
category_counts = df["Category"].value_counts()
plt.pie(category_counts, labels=category_counts.index, autopct="%1.2f%%")
# *autopct = "%1.1f%%" [it means only one decimanl point will be shown]
# *autopct = "%1.2f%%" [it means only two decimanl point will be shown]
plt.title("Category Distribution (Matplotlib)")
plt.show()
```

HISTOGRAM

```python
# Histogram (Growth distribution)
plt.hist(df["Growth"], bins=5, color="purple", edgecolor="black")
plt.title("Growth Distribution (Matplotlib)")
plt.xlabel("Growth Rate")
plt.ylabel("Frequency")
plt.show()
```

Seaborn Visualization

LINE PLOT

```python
sns.lineplot(x="Year", y="Sales", data=df, marker="o", color="blue")
plt.title("Sales Over Years (Seaborn)")
plt.show()
```

BAR CHART

```python
sns.barplot(x="Category", y="Profit", data=df, palette="Set1", estimator="sum")
plt.title("Profit by Year (Seaborn)")
plt.show()
```

SCATTER PLOT

```python
sns.scatterplot(x="Sales", y="Profit", data=df, color="red")
plt.title("Sales vs Profit (Seaborn)")
plt.show()
```

BOX PLOT

```python
sns.boxplot(x="Category", y="Profit", data=df, palette="Set3")
plt.title("Profit Distribution by Category (Seaborn)")
plt.show()
```
